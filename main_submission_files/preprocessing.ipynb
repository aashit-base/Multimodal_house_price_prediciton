## Things to notice in the dataset

- Checking the spread of latitudes and longitudes from data , we can infer that the houses are very close and on checking coordinates , they all are near seattle in the state of Washington,USA.

- From zipcodes range we get that all houses are located in King County, Washington, which encompasses the greater Seattle metropolitan area. We may be able to differentiate the areas into different urban types on the basis of these codes , could be useful.
- Date is not useful to us as yearbuilt and year_renovated does the job for it. Similarly , id is not useful

```python
#Lets use median prices to differentiate the zipcodes into tiers of areas
zip_medians = train_df.groupby('zipcode')['price'].median()
train_df['zip_median_price'] = train_df['zipcode'].map(zip_medians)

# Using binning to separate 5 being highest tier.
train_df['zip_tier'] = pd.qcut(train_df['zip_median_price'], q=5, labels=False) + 1

print(train_df[['zipcode', 'zip_median_price','zip_tier']].head())
```

```python
# HOw many houses are in each tier
print("Count of houses per Tier:")
print(train_df['zip_tier'].value_counts().sort_index())
```

```python
#checking for correlation between price and zip tier
corr_matrix = train_df[['price', 'zip_tier']].corr()
print(corr_matrix)
```

```python
#mapping same column into test dataset
test_df['zip_median_price'] = test_df['zipcode'].map(zip_medians)
#Filling Null values with median
test_df['zip_median_price'] = test_df['zip_median_price'].fillna(train_df['zip_median_price'].median())

#NOw we must apply tiers in test according to train dataset , hence we must use the same tier binning
def tier_making(val):
  # get binning boundaries
  bins = pd.qcut(train_df['zip_median_price'], q=5, retbins=True)[1]
  #sort and return
  return np.digitize(val, bins[1:-1]) + 1

test_df['zip_tier'] = test_df['zip_median_price'].apply(tier_making)
```

## More Feataure Engineering to help fit the data

1. House Age - Older houses may have different valuations
2. Renovation Flag - Indicates if property has been renovated (binary)
3. Total Rooms - Combined bedroom + bathroom count for overall size indicator


```python
train_df['house_age'] = 2025 - train_df['yr_built']
test_df['house_age'] = 2025 - test_df['yr_built']
train_df['was_renovated'] = (train_df['yr_renovated'] > 0).astype(int)
test_df['was_renovated'] = (test_df['yr_renovated'] > 0).astype(int)
```

```python
train_df['total_rooms'] = train_df['bedrooms'] + train_df['bathrooms']
test_df['total_rooms'] = test_df['bedrooms'] + test_df['bathrooms']
```

```python
print("TOP FEATURES CORRELATED WITH LOG OF PRICE:")
corr_with_price = corr_matrix['price_log'].sort_values(ascending=False)
print(corr_with_price.head(15))
```

```python
# FINAL FEATURE SELECTION

# Features to drop:
# - sqft_above: collinear with sqft_living (0.88)
# - total_rooms: collinear with bedrooms/bathrooms (0.84+)
# - zip_median_price: intermediate variable, we have zip_tier now
# - zipcode: replaced by zip_tier (categorical encoding)
# - price: original target (we use price_log)

drop_features = ['sqft_above', 'total_rooms',
                 'zip_median_price', 'zipcode', 'price']
train_df_clean = train_df.drop(columns=[col for col in drop_features if col in train_df.columns])

# Dropping stuff from test
test_drop = ['sqft_above', 'total_rooms', 'zip_median_price', 'zipcode']
test_df_clean = test_df.drop(columns=[col for col in test_drop if col in test_df.columns])

print(f"Train final shape: {train_df_clean.shape}")
print(f"Test final shape: {test_df_clean.shape}")
```

```python
train_df_clean.to_csv('train_preprocessed.csv', index=False)
test_df_clean.to_csv('test_preprocessed.csv', index=False)

print(" Preprocessing Complete !!!")
```
